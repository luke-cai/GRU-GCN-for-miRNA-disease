```python
def labeler(input):
  input_list = input.split("_")
  if input_list[0] != "genetics":
    if input_list[-1] == "down":
      return -1
    elif input_list[-1] == "up":
      return 1
    elif input_list[-1] == "ns":
      return 0
  elif input_list[0] == "genetics":
    if input == "genetics_overexpression_promote":
      return 1
    elif input == "genetics_knock down_promote":
      return -1
    elif input == "genetics_overexpression_suppress":
      return -1
    elif input == "genetics_knock down_suppress":
      return 1


  raise BaseException("Input Error", input)
for i in range(len(df_all)):
  mir = df_all.iloc[i]['mir']
  disease = df_all.iloc[i]['disease']
  category = df_all.iloc[i]['category']
# for i in range(len(df_test)):
#   mir = df_test.iloc[i]['mir']
#   disease = df_test.iloc[i]['disease']
#   category = df_test.iloc[i]['category']
  full_name = mir + '*' + disease
  category = category.replace("-","_")
  category = labeler(category)
  if category == -1:
    df_count.down[df_count.full_name == full_name] += 1  
  elif category == 0:
    df_count.ns[df_count.full_name == full_name] += 1  
  elif category == 1:
    df_count.up[df_count.full_name == full_name] += 1
  else:
    print(full_name)
    raise BaseException("Input Error")
  if df_count.down[df_count.full_name == 'hsa-mir-125a*Breast Neoplasms'].item() >5:
    raise BaseException("Error", i, full_name)
df_count
```

```python
def result(up, down, ns):
  if up > down and up > ns:
    return 'up'
  if down > up and down > ns:
    return 'down'
  if ns > up and ns > down:
    return 'ns'
  if up == down:
    return 'delete'
  if up == ns:
    return 'up'
  if down == ns:
    return 'down'
def result_soft(up, down, ns):
  return (up*1+down*(-1))/(up+down+ns)
re_l = []
re_s = []
total = []
disease = []
mir = []
for i in range(len(df_count)):
  full_name = df_count.iloc[i]['full_name']
  up = df_count.iloc[i]['up']
  down = df_count.iloc[i]['down']
  ns = df_count.iloc[i]['ns']
  total.append(up+down+ns)
  re_l.append(result(up, down, ns))
  re_s.append(result_soft(up, down, ns))
  mir.append(full_name.split('*')[0])
  disease.append(full_name.split('*')[1])
df_count['result'] = re_l
df_count['total'] = total
df_count['disease'] = disease
df_count['mir'] = mir
df_count['result_soft'] = re_s
df_count
```

```python
def re_mature(mir):  
  return list(df_convert[df_convert['mir'] == mir]['seq']), list(df_convert[df_convert['mir'] == mir]['MatureName']), list(df_convert[df_convert['mir'] == mir]['H_seq']), list(df_convert[df_convert['mir'] == mir]['acc'])
re_mature('hsa-mir-1')
```

```python
is_created = False
for i in tqdm(range(len(df_count))):
  mir = df_count.iloc[i]['mir']
  result = df_count.iloc[i]['result']
  disease = df_count.iloc[i]['disease']
  result_soft = df_count.iloc[i]['result_soft']
  seqs, names, H_seqs, accs = re_mature(mir)
  for j in range(len(names)):
    Mature_name = names[j]
    seq = seqs[j]
    H_seq = H_seqs[j]
    acc = accs[j]
    d = {  'Mature_name':[Mature_name],
         'seq':[seq],
         'mir':[mir],
         'result':[result],
         'disease':[disease],
         'result_soft':[result_soft],
         'H_seq': [H_seq],
         'acc': [acc]
         }
    # print(d)
    if is_created == False:
      # print("aa")
      df_complete = pd.DataFrame(data=d)
      is_created = True
    else:
      df_temp = pd.DataFrame(data=d)
      df_complete = df_complete.append(df_temp, ignore_index = True)
df_complete
```

```python
def re_hairpin_data(data):
  lenth = len(data)
  A_amount = 0
  U_amount = 0
  C_amount = 0
  G_amount = 0
  for i in range(len(data)):
    if data[i] == 'A':
      A_amount += 1
    if data[i] == 'U':
      U_amount += 1
    if data[i] == 'C':
      C_amount += 1
    if data[i] == 'G':
      G_amount += 1
  return [lenth, A_amount, U_amount, C_amount, G_amount, (A_amount/lenth), (U_amount/lenth), (C_amount/lenth), (G_amount/lenth)]
re_hairpin_data('CCUUGGAGUAAAGUAGCAGCACAUAAUGGUUUGUGGAUUUUGAAAAGGUGCAGGCCAUAUUGUGCUGCCUCAAAAAUACAAGG')
```

```python
def snd_data(site):
  # print(site)
  response = requests.get(site)
  soup = BeautifulSoup(response.text, "html.parser")
  seqs = soup.find("td",{"id": "stemLoop"}).text.split('\n')
  aa=0
  au=0
  ac=0
  ag=0
  ua=0
  uu=0
  uc=0
  ug=0
  ca=0
  cu=0
  cc=0
  cg=0
  ga=0
  gu=0
  gc=0
  gg=0
  for i in range(len(seqs[1])):

    # print(seqs[2][i],seqs[3][i],seqs[4][i])
    test = seqs[2][i] + seqs[3][i] + seqs[4][i]
    if test == ('a|a'):
      aa += 1
    if test == ('a|u'):
      au += 1
    if test == ('a|c'):
      ac += 1
    if test == ('a|g'):
      ag += 1
    if test == ('u|a'):
      ua += 1
    if test == ('u|u'):
      uu += 1
    if test == ('u|c'):
      uc += 1
    if test == ('u|g'):
      ug += 1
    if test == ('c|a'):
      ca += 1
    if test == ('c|u'):
      cu += 1
    if test == ('c|c'):
      cc += 1
    if test == ('c|g'):
      cg += 1
    if test == ('g|a'):
      ga += 1
    if test == ('g|u'):
      gu += 1
    if test == ('g|c'):
      gc += 1
    if test == ('g|g'):
      gg += 1
  total = aa + au + ac + ag + ua + uu + uc + ug + ca + cu + cc + cg + ga + gu + gc + gg
  # print(aa,'\t',au,'\t',ac,'\t',ag,'\t',ua,'\t',uu,'\t',uc,'\t',ug,'\t',ca,'\t',cu,'\t',cc,'\t',cg,'\t',ga,'\t',gu,'\t',gc,'\t',gg)
  # print(total)
  return[float(aa), float(au), float(ac), float(ag), float(ua), float(uu), float(uc), float(ug), float(ca), float(cu), float(cc), float(cg), float(ga), float(gu), float(gc), float(gg), aa/total, au/total, ac/total, ag/total, ua/total, uu/total, uc/total, ug/total, ca/total, cu/total, cc/total, cg/total, ga/total, gu/total, gc/total, gg/total]
```
